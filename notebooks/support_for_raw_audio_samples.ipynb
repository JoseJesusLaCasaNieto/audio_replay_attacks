# Support for Raw Audio Samples without header

```python
import librosa
import soundfile
import numpy as np

from pymediainfo import MediaInfo
```

```python
def librosa_read(audio_file_path, target_sample_rate=16000):
    # Check if file has audio header
    media_info = MediaInfo.parse(audio_file_path)
    has_header = any(track.track_type == "Audio" for track in media_info.tracks)

    if has_header:
        print("librosa")
        sig, fs = librosa.load(audio_file_path, sr=target_sample_rate)
    else:
        print("soundfile")
        sig, fs = soundfile.read(audio_file_path, samplerate=target_sample_rate, channels=1, subtype="PCM_16", dtype="int16")
        sig = sig.astype(np.float32) / 32768.0
        np.clip(sig, -1.0, 1.0, out=sig)

    return sig, fs
```

```python
def int16_to_float32(x: np.ndarray) -> np.ndarray:
    """
    Convert an int16 PCM numpy array to float32 in [-1.0, 1.0]
    """
    y = x.astype(np.float32) / 32768.0
    np.clip(y, -1.0, 1.0, out=y)
    return y
```
