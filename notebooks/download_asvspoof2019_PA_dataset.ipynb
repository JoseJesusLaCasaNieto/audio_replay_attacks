# Download PA ASVSpoof2019 PA Dataset

## 0. Libraries

```python
import os
import requests
import zipfile
```

## 1. Get setup

```python
# Base path to store the data
DATA_DIR = "../data/raw"  # Ensure this folder exists

# URL of the training set
TRAINING_URL = "https://datashare.ed.ac.uk/bitstream/handle/10283/3336/PA.zip?sequence=4&isAllowed=y"
ARCHIVE_NAME = "ASVspoof2019_PA.zip"
```

## 2. Functions to download and extract data

```python
# Function to download the file
def download_file(url, dest_folder):
    if not os.path.exists(dest_folder):
        os.makedirs(dest_folder)
    
    file_path = os.path.join(dest_folder, ARCHIVE_NAME)
    
    # Download the file if it doesn't exist already
    if not os.path.exists(file_path):
        print(f"Downloading {url}...")
        response = requests.get(url, stream=True)
        with open(file_path, 'wb') as f:
            f.write(response.content)
        print(f"Download complete: {file_path}")
    else:
        print(f"The file already exists at {file_path}")
    
    return file_path
```

```python
# Function to extract the zip file
def extract_zip(archive_path, extract_to):
    with zipfile.ZipFile(archive_path, 'r') as zip_ref:
        zip_ref.extractall(extract_to)
        print(f"Data extracted to {extract_to}")
    
    os.remove(archive_path)
    print(f"File zip {archive_path} removed")
```

## 3. Download and extract the data

```python
# Download and extract the training set
archive_path = download_file(TRAINING_URL, DATA_DIR)
extract_zip(archive_path, DATA_DIR)
```
