# Open and save audios flac

## Unmodified Audio

```python
import librosa
import soundfile as sf

from IPython.display import Audio

# Path to the audio file
in_path = "/mnt/media/fair/audio/replay_attacks/datasets/ASVSpoof2019/ASVspoof2019_PA_train/flac/PA_T_0000001.flac"
out_path = "/mnt/media/fair/audio/replay_attacks/datasets/ASVSpoof2019/ASVspoof2019_PA_train/PA_T_0000001_1.flac"

# Listen to the original audio
Audio(in_path)
```

## Load the audio with Librosa

```python
import warnings

sampling_rate = 16000

try:
    # Load the audio file with soundfile
    print("Loading the audio file with soundfile...")
    audio, source_sampling_rate = sf.read(in_path)
except Exception as e:
    print(f"{e}")
    with warnings.catch_warnings():
        warnings.simplefilter("ignore")
        # Load the audio file with librosa
        print("\nLoading the audio file with librosa...")
        audio, _ = librosa.load(in_path, sr=sampling_rate)

# Reproduce the audio
Audio(audio, rate=sampling_rate)
```

## Save the audio in another path with Soundfile

```python
# 16bit -> PCM_16
print("Saving the audio file with soundfile...")
sf.write(out_path, audio, sampling_rate, format='flac')
print("Audio file saved successfully!")
```

```python
# Load the new audio file
print("Loading the new audio file with soundfile...")
new_audio, new_sr = sf.read(out_path)
Audio(new_audio, rate=new_sr)
```
